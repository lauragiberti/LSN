## Exercise 02.1
in questo esercizio abbiamo calcolato con due metodi montecarlo il risultato del seguente integrale: $I = \int_0^1 f(x) dx=\int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$

i metodi usati sono:
- calcolo dell'integrale usando una distribuzione uniforme in $[0,1]$
- calcolo dell'integrale con il metodo importance sampling (quindi usando una distribuzione non uniforme in $[0,1]$)

### distribuzione uniforme

per calcolare l'integrale:
- ho generato $Nhit=10^4$ numeri casuali ($x$) distribuiti uniformemente in [0,1]
- ho calcolato il risultato dell'integrale nel seguente modo: $I_{10^4 hit}=\sum_{i=1}^{10^4}\frac{f(x_i)}{10^4}$
- ho calcolato l'errore associato nel seguente modo: $errI_{10^4 hit}=\sum_{i=1}^{10^4}\frac{f(x_i)^2}{10^4}-I_{10^4 hit}^2$ per ogni integrale il valore di tale errore era circa 0.23
- ho calcolato in tutto $10^4$ risultati dell'integrale e ne ho fatto la media a blocchi (100 blocchi da 100 componenti)
- ho infine calcolato la media e la varianza progressiva e ne ho fatto un grafico

### importance sampling

prima di calcolare l'integrale ho dovuto trovare una distribuzione di probabilità $p(x)$ che approssimasse $f(x)$. 
la distribuzione che ho scelto è $p(x)=\frac{2}{3}(1-x^2)$

per il calcolo dell'integrale:
- ho generato $Nhit=10^4$ numeri casuali ($x$) distribuiti come $p(x)$ in [0,1], per farlo ho usato il metodo accept reject
- ho calcolato il risultato dell'integrale nel seguente modo: $I_{10^4 hit}=\sum_{i=1}^{10^4}\frac{f(x_i)}{p(x_i)}\frac{1}{10^4}$
- ho calcolato l'errore associato nel seguente modo: $errI_{10^4 hit}=\sum_{i=1}^{10^4}\frac{f(x_i)^2}{p(x_i)^2}\frac{1}{10^4}-I_{10^4 hit}^2$ per ogni integrale il valore di tale errore era circa 0.0024
- ho calcolato in tutto $10^4$ risultati dell'integrale e ne ho fatto la media a blocchi (100 blocchi da 100 componenti)
- ho infine calcolato la media e la varianza progressiva e ne ho fatto un grafico


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
medie_unif=np.loadtxt("I_unif_medie")
var_unif=np.loadtxt("I_unif_var")

medie_imps=np.loadtxt("I_imps_medie")
var_imps=np.loadtxt("I_imps_var")

x=np.arange(0, len(medie_unif),1)
att=np.zeros(len(medie_unif))+1

plt.errorbar(x,medie_unif,yerr=var_unif**0.5, label="misurato")
plt.plot(x,att,color="red", label="atteso")
plt.xlabel('N blocchi')
plt.ylabel("risultato integrale")
plt.grid(True)
plt.legend()
plt.title("distribuzione uniforme")
plt.show()

plt.errorbar(x,medie_imps,yerr=var_imps**0.5, label="misurato")
plt.plot(x,att,color="red", label="atteso")
plt.xlabel('N blocchi')
plt.ylabel("risultato integrale")
plt.grid(True)
plt.legend()
plt.title("importance sampling")
plt.show()
```

## Exercise 02.2
in questo esercizio abbiamo calcolato la deviazione standard in funzione del numero di step per un Random Walk tridimensionale di due tipologie diverse:
- RW 3D con il movimento vincolato ad un reticolo cubico 
- RW 3D nello spazio tridimensionale continuo

per simulare il singolo step del RW sul reticolo cubico:
- ho generato un numero casuale (r) in [0,1]
- ho diviso l'intervallo [0,1] in 6 intervalli uguali (tanti quanti le direzioni possibili)
- ho controllato in che intervallo cadesse il mio numero r e in base a ciò ho scelto la direzione in cui far proseguire il RW 

per simulare il singolo step del RW nello spazio 3D continuo:
- ho generato due numeri casuali theta in [0,$2\pi$] e phi in [0,$\pi$]
- ho trovato le coordinate del RW alla fine dello step   usando la direzione identificata dai due angoli generati

per ogni tipologia di RW:
- ho simulato $10^4$ RW lunghi 100 step, con passo pari a 1 e partenza nell'origine
- per ogni step ho calcolato la media e la deviazione standard delle posizioni

nel limite continuo, quindi facendo simulazioni su una quantità di numeri tendente a infinito, il RW ha un'andamento diffusivo


```python
sigma_cubo=np.loadtxt("sigmaRW_cubo")
sigma_cont=np.loadtxt("sigmaRW_cont")
media_cubo=np.loadtxt("mediaRW_cubo")
media_cont=np.loadtxt("mediaRW_cont")

x=np.arange(0, len(sigma_cubo),1)

plt.errorbar(x,media_cubo,yerr=sigma_cubo)
plt.xlabel('step')
plt.ylabel("$<r>$")
plt.grid(True)
plt.title("$<r>$ RW su reticolo cubico")
plt.show()

plt.plot(x,sigma_cubo, label="osservato")
plt.plot(x,0.47*x**0.5, color="red", label="atteso")
plt.xlabel('step')
plt.ylabel("$\sigma(r)$")
plt.grid(True)
plt.title("$\sigma(r)$ RW su reticolo cubico")
plt.legend()
plt.show()


plt.errorbar(x,media_cont,yerr=sigma_cont)
plt.xlabel('step')
plt.ylabel("$<r>$")
plt.grid(True)
plt.title("$<r>$ RW nello spazio 3D")
plt.show()

plt.plot(x,sigma_cont, label="osservato")
plt.plot(x,0.42*x**0.5, color="red", label="atteso")
plt.xlabel('step')
plt.ylabel("$\sigma(r)$")
plt.grid(True)
plt.title("$\sigma(r)$ RW nello spazio 3D")
plt.legend()
plt.show()
```
