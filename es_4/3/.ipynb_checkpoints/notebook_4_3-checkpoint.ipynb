## ESERCIZIO 4.3

in questo esercizio abbiamo simulato la dinamica molecolare dell'argon nello stato:
- solido
- liquido
- gassoso

per poi determinare la temperatura del sistema ($T$) e l'energia per particella ($Etot/N$, $Ekin/N$ e $Epot/N$)

per fare ciò per ogni stato dell'argon:
- ho termalizzato il sistema facendo 1000 step con rescale delle velocità ogni 10 step
- ho avviato la simulazione di $10^4$ step misurando temperatura ed energia ogni 10 step 
- ho suddiviso i valori in 100 blocchi da 10 valori ciascuno e ne ho calcolato la media a blocchi
- ho calcolato e fatto il grafico della media a blocchi progressiva con il rispettivo errore

inoltre ho effettuato il passaggio dalle unità di misura ridotte alle unità di misura del S.I.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.constants import k as kb

%matplotlib inline

print(kb)

sT=np.loadtxt("T_solid")*120
sT_var=np.loadtxt("T_solid_var")*120**2
sEtot=np.loadtxt("Etot_solid")*120*kb
sEtot_var=np.loadtxt("Etot_solid_var")*(120*kb)**2
sEkin=np.loadtxt("Ekin_solid")*120*kb
sEkin_var=np.loadtxt("Ekin_solid_var")*(120*kb)**2
sEpot=np.loadtxt("Epot_solid")*120*kb
sEpot_var=np.loadtxt("Epot_solid_var")*(120*kb)**2

lT=np.loadtxt("T_liquid")*120
lT_var=np.loadtxt("T_liquid_var")*120**2
lEtot=np.loadtxt("Etot_liquid")*120*kb
lEtot_var=np.loadtxt("Etot_liquid_var")*(120*kb)**2
lEkin=np.loadtxt("Ekin_liquid")*120*kb
lEkin_var=np.loadtxt("Ekin_liquid_var")*(120*kb)**2
lEpot=np.loadtxt("Epot_liquid")*120*kb
lEpot_var=np.loadtxt("Epot_liquid_var")*(120*kb)**2

gT=np.loadtxt("T_gas")*120
gT_var=np.loadtxt("T_gas_var")*120**2
gEtot=np.loadtxt("Etot_gas")*120*kb
gEtot_var=np.loadtxt("Etot_gas_var")*(120*kb)**2
gEkin=np.loadtxt("Ekin_gas")*120*kb
gEkin_var=np.loadtxt("Ekin_gas_var")*(120*kb)**2
gEpot=np.loadtxt("Epot_gas")*120*kb
gEpot_var=np.loadtxt("Epot_gas_var")*(120*kb)**2

x=np.arange(0, len(sT),1)

#Ts=0.8
#Tl=1.1
#Tg=1.2
plt.errorbar(x,sT,yerr=sT_var**0.5, label="T solid")
plt.errorbar(x,lT,yerr=lT_var**0.5, label="T liquid")
plt.errorbar(x,gT,yerr=gT_var**0.5, label="T gas")
plt.xlabel('N blocchi')
plt.ylabel("T [K]")
plt.grid(True)
plt.legend()
plt.title("T")
plt.show()

plt.errorbar(x,sEtot,yerr=sEtot_var**0.5, label="Etot/N solid")
plt.errorbar(x,lEtot,yerr=lEtot_var**0.5, label="Etot/N liquid")
plt.errorbar(x,gEtot,yerr=gEtot_var**0.5, label="Etot/N gas")
plt.xlabel('N blocchi')
plt.ylabel("Etot/N [J]")
plt.grid(True)
plt.legend()
plt.title("Etot/N")
plt.show()

plt.errorbar(x,sEkin,yerr=sEkin_var**0.5, label="Ekin/N solid")
plt.errorbar(x,lEkin,yerr=lEkin_var**0.5, label="Ekin/N liquid")
plt.errorbar(x,gEkin,yerr=gEkin_var**0.5, label="Ekin/N gas")
plt.xlabel('N blocchi')
plt.ylabel("Ekin/N [J]")
plt.grid(True)
plt.legend()
plt.title("Ekin/N")
plt.show()

plt.errorbar(x,sEpot,yerr=sEpot_var**0.5, label="Epot/N solid")
plt.errorbar(x,lEpot,yerr=lEpot_var**0.5, label="Epot/N liquid")
plt.errorbar(x,gEpot,yerr=gEpot_var**0.5, label="Epot/N gas")
plt.xlabel('N blocchi')
plt.ylabel("Epot/N [J]")
plt.grid(True)
plt.legend()
plt.title("Epot/N")
plt.show()
```
