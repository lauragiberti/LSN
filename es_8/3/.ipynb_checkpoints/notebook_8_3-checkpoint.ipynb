in questo esercizio abbiamo cercato il ground state di una particella quantistica in una dimensione sottoposta sempre allo stesso potenziale dell'esercizio precedente ma utilizzando il codice PIGS (Path Integral Ground State).

abbiamo valutato il sistema con due funzioni d'onda di prova diverse:
- $\psi=1$  (costante)
- $\psi=\Psi_T^{\sigma,\mu}(x)$  (la funzione trovata nel precedente esercizio con i parametri gi√† ottimizzati)

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

#phi=1 const

x,p,e = np.loadtxt("prob_const", usecols=(0,1,2), unpack=True)
fig = plt.figure(figsize=(12.8,9.6))
plt.errorbar(x, p, yerr=e)
plt.grid(True)
plt.title("$\psi=1$")
plt.xlabel('x')
plt.show()

n, Ekin, errkin = np.loadtxt("Ekin_const", usecols=(0,1,2), unpack=True)
n, Epot, errpot = np.loadtxt("Epot_const", usecols=(0,1,2), unpack=True)
plt.errorbar(n,Ekin,yerr=errkin, label="$E_{kin}$")
plt.errorbar(n,Epot,yerr=errpot, label="$E_{pot}$")
plt.grid(True)
plt.title("Energia  con  $\psi=1$")
plt.xlabel('time slice')
plt.legend()
plt.show()

#phi=phi(mu,sigma)

x,p,e = np.loadtxt("prob_phi", usecols=(0,1,2), unpack=True)
fig = plt.figure(figsize=(12.8,9.6))
plt.errorbar(x, p, yerr=e)
plt.grid(True)
plt.title("$\psi_{\mu , \sigma}$  $\mu=0.09$ ,  $\sigma=1.03$")
plt.xlabel('x')
plt.show()


n, Ekin, errkin = np.loadtxt("Ekin_phi", usecols=(0,1,2), unpack=True)
n, Epot, errpot = np.loadtxt("Epot_phi", usecols=(0,1,2), unpack=True)
plt.errorbar(n,Ekin,yerr=errkin, label="$E_{kin}$")
plt.errorbar(n,Epot,yerr=errpot, label="$E_{pot}$")
plt.grid(True)
plt.title("Energia  con  $\psi=1$")
plt.xlabel('time slice')
plt.legend()
plt.show()
```
