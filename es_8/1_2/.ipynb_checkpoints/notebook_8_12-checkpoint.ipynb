## ESERCIZIO 8

in questo esercizio abbiamo cercato il grond state di una particella in una dimensione tramite l'ottimizzazione variazionale.
la particella era sottoposta ad un potenziale 
$$V(x) = x^4 - \frac{5}{2}x^2$$

dato che questo modello non è risolvibile analiticamente abbiamo supposto una possibile funzione d'onda della particella in funzione di due parametri $\mu$ e $\sigma$ e variando questi parametri abbiamo cercato di stabilire l'energia del ground state. la funzione d'onda della particella che abbiamo usato è:
$$\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+e^{-\frac{(x+\mu)^2}{2\sigma^2}}$$


### esercizio 8.1

come prima cosa abbiamo fissato i parametri $\mu=1.5$ e $\sigma=1$ di $\Psi_T(x)$ e abbiamo calcolato l'energia media. 
per questo calcolo:
- ho campionato la distribuzione di probabilità $|\Psi_T(x)|^2$ con l'algoritmo di Metropolis in cui ho usato una probabilità tentativo uniforme e una probabilità di accettazione data dal rapporto $\frac{|\Psi(x_{new})|^2}{|\Psi(x)|^2}$
- ho calcolato il valor medio di $\frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}$ distribuito secondo la densità di probabilità campionata $|\Psi_T(x)|^2$ 

di seguito le formule che ho usato per questo calcolo:

$$
\langle {\hat H} \rangle_T = 
\int dx |\Psi_T(x)|^2\frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}=
\int dx |\Psi_T(x)|^2(\frac{{-\frac{1}{2}\frac{\partial^2}{\partial x^2}} \Psi_T^{\sigma,\mu}(x)}{\Psi_T^{\sigma,\mu}(x)}+V(x))
$$

per stimare l'energia ho eseguito $2*10^6$ misure che ho suddiviso in $1000$ blocchi da $200$ componenti di cui ho calcolato la media progressiva e l'errore ad essa associato.
ho inoltre fatto un istogramma in cui mostro la densità di probabilità $|\Psi_T(x)|^2$ campionata

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

H=np.loadtxt("H_smooth")
H_err=np.loadtxt("H_err_smooth")
x_i=np.loadtxt("psi2_x")

x=np.arange(0, len(H),1)

fig,g=plt.subplots(1, 1, figsize=(15, 5))
plt.errorbar(x,H,yerr=H_err)
plt.xlabel('N blocchi')
plt.ylabel("$<H>$")
plt.grid(True)
plt.title("$<H>_{\Psi_{T}}$ con $\mu=1.5$ e $\sigma=1$")
plt.show()

fig,g=plt.subplots(1, 1, figsize=(15, 5))
plt.hist(x_i, bins=np.linspace(-5, 5, num=500),density=True);
plt.xlabel('x')
plt.title("distribuzione $|\Psi_{T}(x)|^2$ con $\mu=1.5$ e $\sigma=1$")
plt.show()
```

### esercizio 8.2 

in questa parte dell'esercizio abbiamo fatto l'ottimizzazione vera e propria ossia abbiamo cercato l'energia minima variando i parametri $\mu$ e $\sigma$.
per fare ciò ho variato i parametri muovendomi su una griglia e per ogni coppia di parametri ho calcolato l'energia media con il metodo spiegato sopra (anche per questi valor medi ho usato lo stesso numero di misure e la media a blocchi).

sono partita da un intervallo ampio ma poco fitto: $\mu$ compreso in $[0 , 4]$ a passi di $0.1$ e $\sigma$ compreso in $[0.1 , 3]$ a passi di $0.1$.
dopo aver visto più o meno dove cadevano i valori minimi di energia ho ristretto e infittito l'intervallo di interesse: $\mu$ compreso in $[0 , 0.25]$ a passi di $0.005$ e $\sigma$ compreso in $[0.95 , 1.1]$ a passi di $0.005$.

```python
H_T= np.loadtxt("H_tentativi", usecols=0)
H_T_err= np.loadtxt("H_tentativi", usecols=1)
mu= np.loadtxt("H_tentativi", usecols=2)
sigma= np.loadtxt("H_tentativi", usecols=3)


import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(sigma, mu, H_T, c=H_T, marker='.', lw=2)
ax.set_xlabel('$\mu$')
ax.set_ylabel('$\sigma$')
ax.set_zlabel('<H>')
ax.view_init(10, 30)
plt.title("$<H>_{\mu , \sigma}$")
plt.show()

fig,g=plt.subplots(1, 1, figsize=(15, 5))
plt.scatter(mu,H_T)
plt.xlabel('$\mu$')
plt.ylabel("$<H>$")
plt.grid(True)
plt.title("$<H>_{\mu}$")
plt.show()

fig,g=plt.subplots(1, 1, figsize=(15, 5))
plt.scatter(sigma,H_T)
plt.xlabel('$\sigma$')
plt.ylabel("$<H>$")
plt.grid(True)
plt.title("$<H>_{\sigma}$")
plt.show()
```

```python
m=H_T[0]
for i in range(len(H_T)):
    if m>H_T[i]:
        m=H_T[i]
        im=i
print("energia minima: ",H_T[im]," +- ",H_T_err[im])
print("mu=",mu[im])
print("sigma=",sigma[im])
```

```python
H_T2= np.loadtxt("H_tentativi2", usecols=0)
H_T_err2= np.loadtxt("H_tentativi2", usecols=1)
mu2= np.loadtxt("H_tentativi2", usecols=2)
sigma2= np.loadtxt("H_tentativi2", usecols=3)


import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(sigma2, mu2, H_T2, c=H_T2, marker='.', lw=2)
ax.set_xlabel('$\mu$')
ax.set_ylabel('$\sigma$')
ax.set_zlabel('<H>')
ax.view_init(10, 30)
plt.title("$<H>_{\mu , \sigma}$")
plt.show()

fig,g=plt.subplots(1, 1, figsize=(15, 5))
plt.scatter(mu2,H_T2)
plt.xlabel('$\mu$')
plt.ylabel("$<H>$")
plt.grid(True)
plt.title("$<H>_{\mu}$")
plt.show()

fig,g=plt.subplots(1, 1, figsize=(15, 5))
plt.scatter(sigma2,H_T2)
plt.xlabel('$\sigma$')
plt.ylabel("$<H>$")
plt.grid(True)
plt.title("$<H>_{\sigma}$")
plt.show()
```

ho quindi trovato che il minimo di energia si ha per $\mu=0.09$ e $\sigma=1.03$

ho poi fissato tali valori e fatto un'ulteriore misura dell'energia media, che ho misurato sempre per $2*10^5$ volte e suddiviso le misure in $1000$ blocchi da $200$ componenti. sotto è riportato il grafico della media progressiva e l'istogramma ottenuto con la distribuzione di probabilità $|\Psi^{\mu=0.045 , \sigma=1.035}(x)|^2$

```python
H=np.loadtxt("H_smooth_GS")
H_err=np.loadtxt("H_err_smooth_GS")
x_i=np.loadtxt("psi2_x_GS")

x=np.arange(0, len(H),1)

fig,g=plt.subplots(1, 1, figsize=(15, 5))
plt.errorbar(x,H,yerr=H_err)
plt.xlabel('N blocchi')
plt.ylabel("$<H>$")
plt.grid(True)
plt.title("$<H>_{\Psi_{T}}$ con $\mu=0.09$ e $\sigma=1.03$")
plt.show()

fig,g=plt.subplots(1, 1, figsize=(15, 5))
plt.hist(x_i, bins=np.linspace(-2.5, 2.5, num=500),density=True);
plt.xlabel('x')
plt.title("distribuzione $|\Psi_{T}(x)|^2$ con $\mu=0.09$ e $\sigma=1.03$")
plt.show()
```
