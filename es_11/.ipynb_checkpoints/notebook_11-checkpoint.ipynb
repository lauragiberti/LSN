## ESERCIZIO 11.2
in questo esercizio abbiato realizzato un fit della funzione: $f(x)=4-3x-2x^2+3x^3$ nell'intervallo $x \in [-1,1]$.

per farlo ci siamo serviti delle reti neurali e, facendo diversi tentativi, la configurazione seguente è quella da cui ho ottenuto i risultati migliori

```python
#LIBRERIE

import numpy as np
import matplotlib
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
from tensorflow.keras import backend, activations
from tensorflow.keras import optimizers, losses, metrics

```

come prima cosa preparo i due dataset:
- il dataset di train che serve a ccreare il modello
- il dataset di validation che serve a verificare il corretto funzionamento del modello

sotto è riportato il grafico dei dati di validazione (che sono soggetti a errore) e dell'andamento teorico ideale

```python
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 1000)
x_valid = np.random.uniform(-1, 1, 100)
x_valid.sort()
y_target = 3*x_valid**3-2*x_valid**2-3*x_valid+4 # ideal (target) linear function

sigma = 0.1
y_train = np.random.normal(3*x_train**3-2*x_train**2-3*x_train+4 , sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(3*x_valid**3-2*x_valid**2-3*x_valid+4, sigma)

plt.plot(x_valid, y_target, label='target')
plt.scatter(x_valid, y_valid, color='r', label='validation data')
plt.legend()
plt.grid(True)
plt.show()
```

poi ho creato la rete neurale usando il modello sequential (che è una rete neurale composta da layer sequenziali). ho provato a variare il numero di layer e di neuroni nei layer, alla fine ho optato per la conformazione seguente.

- input layer: composto da 4 neuroni (specifico che il numero di dati di input è pari a 1)
- 4 hidden layer: composti rispettivamente da 15, 20 e 15 neuroni
- output layer: composto da un neurone solo

tutti i neuroni hanno connessioni di tipo Dense, ossia ogni neurone di un layer è collegato a tutti i neuroni del layer successivo. (ho provato ad utilizzare la connessione Dropout ma ho ottenuto risultati più scarsi).


dopo varie prove ho deciso di usare la funzione di attivazione selu.
altre funzioni di attivazione testate sono: 
- elu che fornisce risultati molto simili a selu
- relu anchessa fornisce risultati simili ma meno accurati
- sigmoid che invece ha fornito risultati sballati e poco accurati

```python
model = Sequential()
model.add(Dense(4, input_shape=(1,), activation="selu"))
model.add(Dense(15, activation="selu"))
model.add(Dense(20, activation='selu'))
model.add(Dense(15, activation='selu'))
model.add(Dense(1, activation='selu'))

model.summary()

x=np.arange(-3,3,0.1)
plt.plot(x, backend.eval(activations.selu(x)))
plt.grid(True); 
plt.title("funzione di attivazione selu")
plt.show()
```

compiliamo poi il modello usando:
- stocastic gradient descend (SGD) come ottimizzatore (ossia come algoritmo di ottimizzazzione per minimizzare la funzione costo)
- mean square error (MSE) come funzione costo da minimizzare con l'ottimizzatore scelto
- mean abslolute error (MAE) come metrica che serve a giudicare le performance del modello (la metrica non viene ottimizzata, fornisce solo un'informazione)

poi fittiamo il modello fornendo:
- i dati di training
- la dimensione dei batches: ossia la dimensione dei sottogruppi in cui vengono suddivisi i dati di training (ho sceldo un divisore del numero di dati forniti perchè ho notato che dava risultati migliori) 
- il numero di epoche: che sono il numero di iterazioni su tutto il data set
- dati di validazione (facoltativi ma danno un'indicazione dell'accuratezza del modello)

```python
model.compile(optimizer='sgd', loss='mse', metrics=['mae'])

history = model.fit(x=x_train, y=y_train, 
          batch_size=25, epochs=30,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

ho poi controllato l'accuratezza del modello valutando la sua risposta rispetto ai dati di validazione.

ho realizzato un grafico con la storia dei valori della funzione costo in funzione delle epoche sia per i dati di train che di validazione.
si osserva che la funzione costo diminuisce nel tempo fino a raggiungere un valore di saturazione.

```python
score = model.evaluate(x_valid, y_valid, batch_size=25, verbose=1)

print('Test loss:', score[0])
print('Test accuracy:', score[1])

# summarize history for loss
plt.plot(history.history['loss'], label="train")
plt.plot(history.history['val_loss'], label="validation")
plt.ylabel('model loss')
plt.xlabel('epoch')
plt.legend(loc='best')
plt.show()
```

ho poi fornito al modello dei dati nello stesso intervallo in cui ho realizzato il fit e anche al di fuori e ho comparato la risposta ottenuta con il valore teorico atteso.
si può notare una buona corrispondenza tra dati ottenuti dal modello e i dati attesi all'interno dell'intervallo del fit.
invece per quanto riguarda i dati ottenuti al di fuori del fit non c'è corrispondenza tra i dati ottenuti dal modello e quelli attesi

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.title("dati interpolati")
plt.show()

x_predicted = np.random.uniform(-1.5, 1.5, 100)
y_predicted = model.predict(x_predicted)
x_new=np.random.uniform(-1.5,1.5,100)
x_new.sort()
y_new=3*x_new**3-2*x_new**2-3*x_new+4
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_new, y_new)
plt.grid(True)
plt.title("dati estrapolati")
plt.show()
```

## ESERCIZIO 11.3

in questo esercizio abbiato realizzato un fit della funzione da $R^2 -> R$ : $f(x,y) = \sin(x^2+y^2)$ nell'intervallo $x \in [-3/2,3/2]$ e $y \in [-3/2,3/2]$.


ho tentato di realizzare il fit modificando i parametri ma ho ottenuto scarsi risultati.

```python
import math

xy_valid=3.*np.random.rand(1000,2)-1.5
xy_train=3.*np.random.rand(10000,2)-1.5

sigma = 0.0
f_target=np.zeros(len(xy_valid[:,0]))
f_valid=np.zeros(len(xy_valid[:,0]))
f_train=np.zeros(len(xy_train[:,0]))

for i in range(len(xy_valid[:,0])):
    f_target[i] = math.sin(xy_valid[i,0]**2+xy_valid[i,1]**2)
    f_valid[i] = np.random.normal(math.sin(xy_valid[i,0]**2+xy_valid[i,1]**2), sigma)
               
for i in range(len(x_train)):
    f_train[i] = np.random.normal(math.sin(xy_train[i,0]**2+xy_train[i,1]**2) , sigma)


from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
#ax.scatter(xy_train[:,0], xy_train[:,1], f_train, c=f_train, marker='x', lw=2)
ax.scatter(xy_valid[:,0], xy_valid[:,1], f_target, c=f_target, marker='.', lw=5, label="target")
ax.scatter(xy_valid[:,0], xy_valid[:,1], f_valid, c=f_valid, marker='x', lw=1, label="valid")
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$f(x,y)$')
ax.view_init(10, 30)
ax.legend()
plt.show()
```

```python
model = Sequential()
model.add(Dense(200, input_shape=(2,), activation="selu"))
model.add(Dense(100, activation="selu"))
model.add(Dense(100, activation="selu"))
model.add(Dropout(0.5))
model.add(Dense(100, activation='selu'))
model.add(Dense(50, activation="selu"))
model.add(Dropout(0.5))
model.add(Dense(1, activation='selu'))


model.summary()


model.compile(optimizer='adadelta', loss='mse', metrics=['mse'])

history = model.fit(x=xy_train, y=f_train, 
          batch_size=50, epochs=100,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(xy_valid, f_valid))


score = model.evaluate(xy_valid, f_valid, batch_size=50, verbose=1)

print('Test loss:', score[0])
print('Test accuracy:', score[1])

# summarize history for loss
plt.plot(history.history['loss'], label="train")
plt.plot(history.history['val_loss'], label="validation")
plt.ylabel('model loss')
plt.xlabel('epoch')
plt.legend(loc='best')
plt.show()
```

```python
xy_predicted = 3.*np.random.rand(1000,2)-1.5
f_predicted = model.predict(xy_predicted)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(xy_valid[:,0], xy_valid[:,1], f_target, c=f_target, marker='.', lw=5, label="target")
ax.scatter(xy_predicted[:,0], xy_predicted[:,1], f_predicted, c=f_predicted, marker='x', lw=1, label="predicted")
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$f(x,y)$')
ax.view_init(10, 30)
ax.legend()
plt.show()
```
