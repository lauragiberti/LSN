in questo esercizio esercizio abbiamo completato l'algoritmo del modello di ising in 1D e fatto una simulazione a diverse temperature.

il modello di ising è un modello per la magnetizzazione dei materiali in presenza o meno di un campo magnetico. l'hamiltoniana del modello 1D è:
$$H =-J\sum_{i=1}^N s_i s_{i+1}-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1})$$
in cui:
- il primo termine rappresenta l'interazione tra gli spin ($J$ è il coefficiente di interazione). si ha interazione solo tra i primi vicini, che in questo caso sono solo i due spin adiacenti sotto le periodic boudary condition (quindi l'ultimo spin e il primo sono considerati primi vicini)
- il secondo termine rappresenta invece l'interazione degli spin con il campo magnetico $h$

prima di avviare la simulazione abbiamo completato il codice:
- ho aggiunto una nuova funzione di input che permette di ripartire dalla configurazione finale di una precedente simulazione. ciò sarà utile in seguito per avviare la simulazione dopo un periodo di equilibrazione iniziale
- oltre al calcolo dell'energia del sistema che era già implementato, ho aggiunto la possibilità di calcolare:
 - la capacità termica: $C(N,T) = \beta^2 (\langle H^2 \rangle_{h=0}-\langle H \rangle_{h=0}^2)$
 - la magnetizzazione: $M(N,T,h=0.02)= \langle \sum_{i=1,N}s_i \rangle_{h=0.02}$
 - la sucettività magnetica: $\chi(N,T) = \beta \langle (\sum_{i=1,N}s_i)^2 \rangle_{h=0}$
- ho poi implementato due algoritmi per l'evoluzione del sistema
 - algoritmo di Metropolis: in cui la probabilità di transizione è data dalla probabilità di tentativo (che in questo caso è $1$ perchè modifico in ogni caso lo spin della particella) per la propabilità di accettazione ($min(1, e^{-\beta (E_{new}-E_{old})})$)
 - algoritmo di Gibbs: che è un caso particolare dell'algoritmo di Metropolis in cui pa probabilità di accettazione è pari a $1$. quindi la probabilità di transizione è data direttamente dalla probabilità tentativo che è una probabilità condizionata ($p_{up}=\frac{e^{-\beta E_{up}}}{e^{-\beta E_{up}}+e^{-\beta E_{down}}}$ e $p_{down}=1-p_{up}$)

prima di iniziare con la simulazione vera e propria ho osservato l'andamento dell'energia nei primi $1000$ step alla temperatura di $0.5$, che è la temperatura più bassa che campioneremo.
ho fatto ciò per capire quanti step usare per equilibrare il sistema e quanti step mettere all'interno di ciascun blocco per avere delle misure scorrelate.
ho deciso di prendere la temperatura più bassa perchè è quella che presenta più criticità: infatti partendo da una configurazione iniziale caotica, il sistema a temperatura bassa è quello che ci mette più tempo ad equilibrarsi e inoltre è il sistema che presenta anche più correlazione rispetto a sistemi con temperature superiori.

osservando gli andamenti ho deciso di utilizzare $200$ step per equilibrare il sistema e $200$ step per ogni blocco in modo da ottenere misure scorrelate

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

energy_eq_met=np.loadtxt("energy_eq_met")
energy_eq_gib=np.loadtxt("energy_eq_gib")
x=np.arange(0, len(energy_eq_met),1)

fig,g=plt.subplots(1, 1, figsize=(15, 5))
plt.plot(x,energy_eq_met, label="Metropolis")
plt.plot(x,energy_eq_gib, label="Gibss")
plt.xlabel('step')
plt.ylabel("$E$")
plt.grid(True)
plt.title("$E(T=0.5)$")
plt.legend()
plt.show()
```

abbiamo poi fatto partire la simulazione vera e propria durante cui abbiamo calcolato il valore delle seguenti grandezze per particella al variare della temperatura:
- energia interna in assenza di campo magnetico ($h=0$)
- capacità termica in assenza di campo magnetico ($h=0$)
- sucettività magnetica in assenza di campo magnetico ($h=0$)
- magnetizzazione in presenza di campo magnetico ($h=0.02$)

le temperature per cui abbiamo effettuato le misure sono comprese nell'intervallo $[0.5, 2.0]$ e abbiamo campionato ogni $0.1$.
abbiamo eseguito la stessa simulazione sia con gli step implementati con l'algoritmo di Metropolis sia con l'algoritmo di Gibbs.
è stata effettuata una simulazione a parte per la misurazione della magnetizzazione in quanto è l'unica quantità che richiedeva un campo magnetico non nullo.

prima di misurare le quantità elencate ho eseguito $200$ step a vuoto per equilibrare il sistema. 
poi ho iniziato la misurazione delle quantità: 
- per ogni temperatura e tipologia di step, ho eseguito $10^5$ step, misurando le 4 quantità ad ogni step
- ho poi suddiviso le misure in $500$ blocchi da $200$ componenti ciascuno e ne ho calcolato la media. in questo modo ho ottenuto $500$ valori per ogni quantità piuttosto scorrelati tra loro
- ho quindi potuto fare la media e la deviazione standard della media dai valori della media a blocchi, ottenendo infine un unico valore delle quantità con il rispettivo errore per ogni temperatura.  

```python
en_M=np.loadtxt("en_M")
en_err_M=np.loadtxt("en_err_M")
ct_M=np.loadtxt("ct_M")
ct_err_M=np.loadtxt("ct_err_M")
sm_M=np.loadtxt("sm_M")
sm_err_M=np.loadtxt("sm_err_M")
ma_M=np.loadtxt("ma_M")
ma_err_M=np.loadtxt("ma_err_M")

en_G=np.loadtxt("en_G")
en_err_G=np.loadtxt("en_err_G")
ct_G=np.loadtxt("ct_G")
ct_err_G=np.loadtxt("ct_err_G")
sm_G=np.loadtxt("sm_G")
sm_err_G=np.loadtxt("sm_err_G")
ma_G=np.loadtxt("ma_G")
ma_err_G=np.loadtxt("ma_err_G")

x=np.arange(0.5, 2.1 ,0.1)

plt.errorbar(x,en_M,yerr=en_err_M, label="Metropolis")
plt.errorbar(x,en_G,yerr=en_err_G, label="Gibss")
plt.xlabel('T')
plt.ylabel("$E$")
plt.grid(True)
plt.title("$E(T)$")
plt.legend()
plt.show()

plt.errorbar(x,ct_M,yerr=ct_err_M, label="Metropolis")
plt.errorbar(x,ct_G,yerr=ct_err_G, label="Gibss")
plt.xlabel('T')
plt.ylabel("$C$")
plt.grid(True)
plt.title("$C(T)$")
plt.legend()
plt.show()

plt.errorbar(x,sm_M,yerr=sm_err_M, label="Metropolis")
plt.errorbar(x,sm_G,yerr=sm_err_G, label="Gibss")
plt.xlabel('T')
plt.ylabel("$\chi$")
plt.grid(True)
plt.title("$\chi(T)$")
plt.legend()
plt.show()

plt.errorbar(x,ma_M,yerr=ma_err_M, label="Metropolis")
plt.errorbar(x,ma_G,yerr=ma_err_G, label="Gibss")
plt.xlabel('T')
plt.ylabel("$M$")
plt.grid(True)
plt.title("$M(T)$")
plt.legend()
plt.show()
```
