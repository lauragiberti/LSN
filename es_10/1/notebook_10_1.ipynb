in questo esercizio abbiamo risolto il problema del commesso viaggiatore usando il metodo del simulated annealing.
l'approccio di simulated annealing è una tecnica di ottimizzazione che prende spunto dall'evento fisico del raffreddamento di un sistema.
in pratica si fa evolvere un sistema facendogli sperimentare diverse configurazioni e lo si raffredda poco a poco ossia si diminuisce la probabilità di accettazione in modo da trovare un minimo della funzione costo.

il Simulated Annealing applicato al problema del commesso viaggiatore consiste nel creare un cammino iniziale e farlo evolvere sfruttando delle mutazioni (in particolare sono state usate le mutazioni precedentemente scritte nell'esercitazione 9).
per far evolvere il sistema si usa un algoritmo di Metropolis:
- parto da uno specifico cammino e ne calcolo la lunghezza ($L$)
- applico le mutazioni al mio cammino di partenza ottenendo un nuovo cammino di cui calcolo la lunghezza ($L_{new}$)
- accetto il nuovo cammino con probabilità:
    - $1$ se $L_{new}<L$
    - $e^{- \beta (L_{new}-L)}$ se $L_{new}>L$
il sistema evolve per un certo numero di step a temperatura fissata e poco a poco si diminuisce la temperatura, diminuendo quindi la probabilità di accettazione nel caso $L_{new}>L$.
in tal modo per temperature alte il sistema è in grado di esplorare anche punti lontani invece, al diminuire della temperatura, rimane sempre più localizzato.
questo approccio è utile in quanto aumenta le probabilità di trovare un minino globale.



per risolvere questo problema creato la classe $SA$ (molto simile alla classe $GA$ dell'esercizio precedente)

- come prima cosa ho usato il costruttore della classe che fissa i parametri che verranno usati in seguito $SA(Nc, Tstart, deltaT, Tstep, step)$ in cui:
    - $Nc$ è il numero di città tra cui si vuole trovare il percorso più breve ($Nc=32$)
    - $Tstart$ è la temperatura da cui parte il sistema, ($Tstart=0$). NB nel codice quando si fa riferimento a $T$ in realtà è $\beta =1/T$, quindi per raffreddare il sistema T aumenterà.
    - $deltaT$ è l'aumento di $T$ ($\beta$) ad ogni step ($deltaT=0.5$)
    - $Tstep$ sono il numero di step di temperatura in cui si fa evolvere il sistema ($Tstep=200$)
    - $step$ sono il numero di step a $T$ fissata in cui si fa evolvere il sistema ($step=10000$)


- in seguito ho creato le configurazioni delle città con il metodo $GenerateCities(mode)$
    - se $mode=0$ vengono create $Nc$ città vincolate a una circonferenza di raggio unitario centrata nell'origine
    - se $mode=1$ vengono create $Nc$ città all'interno di un quadrato di lato unitario centrato nell'origine


- ho poi creato il primo cammino con il metodo $GenerateCammino()$. questo metodo crea una permutazione casuale delle città che sono state create con il vincolo di tenere fissa la prima città, che è il punto di arrivo e di partenza (si fa per diminuire la degenerazione dei cammini più brevi che possono essere creati e quindi rendere l'algoritmo più performante)


- ora è tutto pronto per far partire la simulazione con il metodo $Evolution()$. questo metodo svolge l'evoluzione del simulated annealing descritta sopra quindi:
    - la temperatura ($\beta$) viene fatta variare da $0$ a $100$ in $200$ step 
    - ad ogni temperatura ($\beta$) fissata vengono svolti $10^4$ step in cui:
        - al cammino di partenza si applicano le 4 mutazioni già descritte nel notebook 9 
        - il nuovo cammino viene accettato con la probabilità indicata sopra 
        - ogni volta si tiene traccia del cammino più corto (che non per forza coincide con l'ultimo cammino generato)


nei grafici sotto sono mostrati per ogni configurazione delle città:
- la raffigurazione del cammino più corto trovato durante tutta la simulazione
- l'andamento della lunghezza del cammino migliore in funzione degli step di temperatura ($\beta$)

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

#CIRCONFERENZA
xy = np.loadtxt("city_1", delimiter=" " , usecols=(0,1))
bestp=np.loadtxt("bestcammino_1", dtype=int)
LbestT=np.loadtxt("L_best_1", delimiter=" ", usecols=(0,1))

x=xy[:,0]
y=xy[:,1]
Lbest=LbestT[:,0]
T=LbestT[:,1]

xb=np.zeros(len(bestp))
yb=np.zeros(len(bestp))
for i in range(len(bestp)):
    xb[i]=x[bestp[i]]
    yb[i]=y[bestp[i]] 
    
fig,g=plt.subplots(1, 1, figsize=(10, 10))
plt.scatter(x,y)
plt.plot(xb,yb)
plt.xlabel('$x$')
plt.ylabel("$y$")
plt.title("best path circonferenza")
plt.show()

fig,g=plt.subplots(1, 1, figsize=(15, 10))
plt.plot(T,Lbest)
plt.xlabel("$ 1/T $ step")
plt.ylabel("best $L$")
plt.title("best L path circonferenza")
plt.grid(True)
plt.show()

#QUARATO
xy = np.loadtxt("city_2", delimiter=" " , usecols=(0,1))
bestp=np.loadtxt("bestcammino_2", dtype=int)
LbestT=np.loadtxt("L_best_2", delimiter=" ", usecols=(0,1))

x=xy[:,0]
y=xy[:,1]
Lbest=LbestT[:,0]
T=LbestT[:,1]

xb=np.zeros(len(bestp))
yb=np.zeros(len(bestp))
for i in range(len(bestp)):
    xb[i]=x[bestp[i]]
    yb[i]=y[bestp[i]] 
    
fig,g=plt.subplots(1, 1, figsize=(10, 10))
plt.scatter(x,y)
plt.plot(xb,yb)
plt.xlabel('$x$')
plt.ylabel("$y$")
plt.title("best path quadrato")
plt.show()

fig,g=plt.subplots(1, 1, figsize=(15, 10))
plt.plot(T,Lbest)
plt.xlabel("$ 1/T $ step")
plt.ylabel("best $L$")
plt.title("best L path quadrato")
plt.grid(True)
plt.show()
```
