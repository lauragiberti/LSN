## ESERCIZIO 7.3 - 7.4

in questo esercizio abbiamo aggiunto il calcolo della funzione di distribuzione radiare $g(r)$ e della pressione $P$ al codice della dinamica molecolare implementato con l'algoritmo di Verlet.
l'obiettivo di questo esercizio è quello di dimostrare che, nonostante l'uso di due algoritmi basati su funzionamenti diversi, si ottengono comunque risultati analoghi. infatti:
- la dinamica molecolare con l'algoritmo di Verlet sfrutta le equazioni del moto per far progredire il sistema
- la dinamica molecolare con l'algoritmo di Metropolis invece è di natura puramente stocastica

dopo aver inserito il calcolo della $g(r)$, ho fatto partire la simulazione:
- come prima cosa ho termalizzato il sistema facendo $10^3$ step con rescale delle velocità ogni $10$ step (in questa parte della simulazione non sono state eseguite le misure)
- ho poi avviato la simulazione vera e propria facendo $2*10^4$ step suddivisi in $100$ blocchi da $200$ componenti ciascuno. ad ogni step ho eseguito la misura di:
  - energia per particella ($E_{tot}$))
  - pressione ($P$)
  - funzione di distribuzione radiale ($g(r)$)
  
sotto sono riportati i risultati che, se confrontati con i grafici del precedente esercizio, risultano analoghi. 


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.constants import k as kb

%matplotlib inline

#print(kb)

etot_S=np.loadtxt("etot_S")*120*kb
etot_err_S=np.loadtxt("etot_err_S")*120*kb
pres_S=np.loadtxt("pres_S")*120*kb/(0.34*10**(-12))
pres_err_S=np.loadtxt("pres_err_S")*120*kb/(0.34*10**(-12))
gr_S=np.loadtxt("gr_S")
gr_err_S=np.loadtxt("gr_err_S")

etot_L=np.loadtxt("etot_S")*120*kb
etot_err_L=np.loadtxt("etot_err_L")*120*kb
pres_L=np.loadtxt("pres_L")*120*kb/(0.34*10**(-12))
pres_err_L=np.loadtxt("pres_err_L")*120*kb/(0.34*10**(-12))
gr_L=np.loadtxt("gr_L")
gr_err_L=np.loadtxt("gr_err_L")

etot_G=np.loadtxt("etot_G")*120*kb
etot_err_G=np.loadtxt("etot_err_G")*120*kb
pres_G=np.loadtxt("pres_G")*120*kb/(0.34*10**(-12))
pres_err_G=np.loadtxt("pres_err_G")*120*kb/(0.34*10**(-12))
gr_G=np.loadtxt("gr_G")
gr_err_G=np.loadtxt("gr_err_G")

x1=np.arange(0, len(etot_S),1)
x2=np.arange(0, len(gr_S),1)

fig,g=plt.subplots(1, 1, figsize=(15, 5))
plt.errorbar(x1,etot_S,yerr=etot_err_S, label="solid")
plt.errorbar(x1,etot_L,yerr=etot_err_L, label="liquid")
plt.errorbar(x1,etot_G,yerr=etot_err_G, label="gas")
plt.xlabel('N blocchi')
plt.ylabel("E [J]")
plt.grid(True)
plt.legend()
plt.title("$E_{tot}$")
plt.show()

fig,g=plt.subplots(1, 1, figsize=(15, 5))
plt.errorbar(x1,pres_S,yerr=pres_err_S, label="solid")
plt.errorbar(x1,pres_L,yerr=pres_err_L, label="liquid")
plt.errorbar(x1,pres_G,yerr=pres_err_G, label="gas")
plt.xlabel('N blocchi')
plt.ylabel("P [Pa]")
plt.grid(True)
plt.legend()
plt.title("$P$")
plt.show()

fig,g=plt.subplots(1, 1, figsize=(15, 5))
plt.errorbar(x2,gr_S,yerr=gr_err_S, label="solid")
plt.errorbar(x2,gr_L,yerr=gr_err_L, label="liquid")
plt.errorbar(x2,gr_G,yerr=gr_err_G, label="gas")
plt.xlabel('N blocchi')
plt.ylabel("g")
plt.grid(True)
plt.legend()
plt.title("$g(r)$")
plt.show()
```
