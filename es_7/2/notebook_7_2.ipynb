in questa parte dell'esercizio abbiamo completato il codice della dinamica molecolare implementato con l'algoritmo di metropolis, aggiungendo la possibilità di calcolare la funzione di distribuzione radiale delle particelle $g(r)$.

la funzione di distribuzione radiale è definita come la probabilità di trovare due particelle a distanza $r$ in un determinato gas, rispetto alla probabilità di trovarle alla stessa distanza in un gas ideale, quindi senza potenziale di interazione.
è interessante riuscire a stimare questa grandezza perchè ci permette di calcolare il valor medio di grandezze che dipendono dalla posizione ralativa di due corpi, come ad esempio il potenziale.

la $g(r)$ è stata calcolata creando un istogramma in cui le distanze comprese in $[0,L/2]$ sono state suddivise in $100$ bin, ogni bin è stato aumentato di due per ogni coppia con distanza compresa nel bin. ho poi normalizzato dividendo per $\rho N \Delta V(r)$ in cui $\rho$ è la densità, $N$ è il numero di particelle e $\Delta V(r)$ è la differenza del potenziale tra i bin.

ho poi avviato la simulazione per i 3 stati: solido, liquido e gassoso.
in ogni simulazione ho:
- fatto eseguire $200$ step per equilibrare il sistema (senza misurare le quantità di interesse)
- poi ho fatto partire la simulazione vera e propria in cui ho misurato la $g(r)$ per $2*10^4$ step suddivisi in $100$ blocchi da $200$ componenti (il numero di cmponenti per blocco è stato deciso osservando i grafici dell'andamento dell'errore in funzione della dimensione dei blocchi nell'esercizio precedente)

si osserva che il nostro sistema si comporta in modo diverso rispetto ad un gas perfetto, infatti le particelle non si distribuiscono con distanze del tutto casuali ma ci sono alcune distanze che sono più gettonate.
infatti la funzione $g(r)$ presenta dei picchi in corrispondenza di alcune distanze.
inoltre si nota che non sono presenti coppie di particelle con una distanza inferiore ad un determinato valore che cambia in base allo stato del sistema. 


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

gr_S=np.loadtxt("gr_S")
err_S=np.loadtxt("gr_err_S")
gr_L=np.loadtxt("gr_L")
err_L=np.loadtxt("gr_err_L")
gr_G=np.loadtxt("gr_G")
err_G=np.loadtxt("gr_err_G")

x=np.arange(1/200, len(gr_S)/100,1/100)

fig,g=plt.subplots(1, 1, figsize=(15, 5))
plt.errorbar(x,gr_S,yerr=err_S, label="solido")
plt.xlabel('distanza $[0,L/2]$')
plt.ylabel("coppie")
plt.grid(True)
plt.title("$g(r)$ solido")
plt.show()

fig,g=plt.subplots(1, 1, figsize=(15, 5))
plt.errorbar(x,gr_L,yerr=err_L, label="liquido")
plt.xlabel('distanza $[0,L/2]$')
plt.ylabel("coppie")
plt.grid(True)
plt.title("$g(r)$ liquido")
plt.show()

fig,g=plt.subplots(1, 1, figsize=(15, 5))
plt.errorbar(x,gr_G,yerr=err_G, label="gassoso")
plt.xlabel('distanza $[0,L/2]$')
plt.ylabel("coppie")
plt.grid(True)
plt.title("$g(r)$ gassoso")
plt.show()
```
